1.Linear Interpolation

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#interpolate the  data using linear interpolation
x_new=np.linspace(1,5,10)#crete new x values
y_interp=np.interp(x_new,x,y)#interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

2.cubic interpolation  with scipy

```python
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
#crete a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
#interpolation the data
x_new=np.linspace(1,5,10)
y_interp=f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

3.polynomail interpolation

```python
import numpy as np

#crate some  sample data
x=np.array([1,2,3,4,5])
y=np.array([1,4,27,64,125])
```

```python
#interpolate  the data usng polynomail interpolation
p=np.polyfit(x,y,4) #fit a 2nd degree  polynomial to the data
```

```python
x_new=np.linspace(1,5,10)#crete a x new value
y_interp=np.polyval(p,x_new)#interpolate the y value
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
