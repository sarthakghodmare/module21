```python
from sklearn.datasets import make_classification#skelearn uses for machine learning alogorithm
```

```python
#x is independent feature
#y is dependent feature it is specific output
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
x#it has 2 feature
```

```python
y#output of above specific features
```

```python
#convet above into dataframe
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()#purely imbalanced dataset
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```
