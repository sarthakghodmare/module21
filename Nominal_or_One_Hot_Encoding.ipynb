##Nominal or One Hot Encoding

```python
import pandas as pd
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#creata a  sample  dataframe  with a categorical variable
df=pd.DataFrame({
    'color':['red','blue','green','green','red','blue']
})
```

```python
df
```

```python
#creat an instance of an onehot encoder
encoder=OneHotEncoder()
```

```python
#convert in array
encoder.fit_transform(df[['color']]).toarray()#sort in alaphabetically value first blue ,green ,red
```

```python
#fit the  encoder to the dataframe  and transform  the categorical  variable
encoded=encoder.fit_transform(df[['color']])
```

```python
encoded
```

```python
import pandas as pd
encoded_df=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
columns=encoder.get_feature_names_out()
```

```python
columns
```

```python
encoded_df
```

```python
#append with df
pd.concat([df,encoded_df],axis=1)
```
